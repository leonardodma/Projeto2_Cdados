# Separação dos Dados:

```python
# Importanto Bibliotecas para o Trabalho:
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import statsmodels.api as sm

import scipy as scp
import sklearn

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn import metrics
from sklearn.metrics import confusion_matrix
```

```python
dados = pd.read_excel('classificacao_musicas_leonardo.xlsx')
dados.head(5)
```

```python
dados['like'].value_counts()
```

```python
hate = dados.loc[dados['like'] == 0, :]
like = dados.loc[dados['like'] == 1, :]
love = dados.loc[dados['like'] == 2, :]
```

# Criando Gráficos Comparativos 

```python
# Funçção para cirar histogramas em loop
def Hist(variavel, df1, df2, df3):
    plt.style.use('bmh')
    sns.distplot(df1[variavel], color="darksalmon", label="Hate")
    sns.distplot(df2[variavel], color="skyblue", label="Like")
    sns.distplot(df3[variavel], color="lightgreen", label="Love")
    plt.legend(loc='best')
    plt.ylabel('densidade')
    plt.xlabel(variavel)


features = [
    'danceability', 'energy', 'key', 'loudness', 'speechiness', 'acousticness',
    'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms'
]

plt.figure(figsize=(20, 40))

for i in range(len(features)):
    plt.subplot(6, 2, i + 1)
    Hist(features[i], hate, like, love)
```

# Considerações do Modelo 

Por meio da análise dos histogramas, foi possível concluir que as varáveis que mais tiveram divergência foram:
* danceability
* energy
* loudness
* speechiness
* liveness
* tempo

Assim, essas srão as variáveis que serão utilizadas para a modelagem e aprofundamento do classificador.

# Regressão Logística Multinomial

#### STATSMODELS

```python
x_train = dados.drop([
    'like', 'name', 'key', 'acousticness', 'instrumentalness', 'valence', 'id',
    'uri', 'track_href', 'analysis_url', 'time_signature', 'mode',
    'duration_ms'
],
                     axis=1)

y_train = dados['like']
x_train.head(3)
```

```python
# Removendo dados por meio do valor p:
logit_model = sm.MNLogit(y_train, sm.add_constant(x_train))
logit_model
result = logit_model.fit()
stats1 = result.summary()
stats2 = result.summary2()
print(stats1)
print(stats2)
```

*Considerações*:
* O tempo demonstrou não ser uma boa variável para a modelagem, pois apresenta mais que 50% de erro em todas as tabelas comparativas.
* Apesar da variável "energia" ser um bom comparativo das músicas menos gostadas com relção as músicas amadas, ela é muito ruim para relacionar músicas curtidas com as músicas amadas.
